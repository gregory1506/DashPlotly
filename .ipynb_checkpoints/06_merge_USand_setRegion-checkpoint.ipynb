```python
countries = df7["Country/Region"].unique()
URL = "https://restcountries.eu/rest/v2/name/"
cinfo = {}
for country in countries:
    tmp_country = ""
    if country == "Congo (Brazzaville)":
        tmp_country = "Congo-Brazzaville"
    elif country == "Congo (Kinshasa)":
        tmp_country = "Congo-Kinshasa"
    elif country == "Czechia":
        tmp_country =  "Czech Republic"
    elif country in ["Diamond Princess","MS Zaandam"]:
        continue
    elif country == "Korea, South":
        tmp_country = "Korea (Republic of)"
    elif country == "North Macedonia":
        tmp_country = "Republic of Macedonia"
    elif country == "Taiwan*":
        tmp_country = "Taiwan"
    elif country == "West Bank and Gaza":
        tmp_country = "Palestine"
    elif country == "US":
        tmp_country = "United States of America"
    else:
        tmp_country = country
    try:
        r =rq.get(url = URL + tmp_country).json()
        population = r[0]['population']
        region = r[0]['region']
        subregion = r[0]['subregion'].replace(" ","_")
        code = r[0]['alpha3Code']
        if country in cinfo:
            continue
        else:
            cinfo[country] = [population,region,subregion,code]
    except:
        print(f'problem with {country}')
cinfo
```

```python
with open("test.json","w") as w:
    json.dump(cinfo,w)
```
